```python
#4
def is_anagram_from_file(path: str) -> bool:
    with open(path, 'r', encoding='utf-8') as f:
        lines = [line.strip() for line in f if line.strip()]
    candidates = []
    for line in lines[-3:]: 
        if line.upper() == "TRUE":
            continue
        candidates.append(line)
    
    if len(candidates) < 2:
        return False
    str1 = candidates[0].replace('|', '').strip()
    str2 = candidates[1].replace('|', '').strip()
    s1 = ''.join(str1.lower().split())
    s2 = ''.join(str2.lower().split())
    
    return sorted(s1) == sorted(s2)
is_anagram_from_file(r"C:\Users\ADMIN\OneDrive\Desktop\Tom Marvolo Riddle.txt")
```

```python
#5
def count(path: str, character: str) -> int:
    try:
        with open(path, 'r', encoding='utf-8') as file:
            sentence = file.read()
        return sentence.count(character)
    except FileNotFoundError:
        print(f"File '{path}' không tồn tại.")
        return 0
    except Exception as e:
        print(f"Có lỗi xảy ra: {e}")
        return 0
count(r"C:\Users\ADMIN\OneDrive\Desktop\count.txt", "e")
```

```python
#6
def countCharacter(path: str, letter: str) -> int:
    try:
        with open(path, 'r', encoding='utf-8') as file:
            content = file.read()
            char = letter[0].lower()
            return content.lower().count(char)
    except FileNotFoundError:
        print(f"File '{path}' không tồn tại.")
        return 0
    except Exception as e:
        print(f"Lỗi: {e}")
        return 0
countCharacter(r"C:\Users\ADMIN\OneDrive\Desktop\count1.txt",'e')
```

```python
#10
def getMoney(path: str) -> list[int]:
    with open(path, 'r', encoding='utf-8') as f:
        content = f.read().strip()
        if not content:
            return []
        return list(map(int, content.split()))

def rob(houses: list[int]) -> int:
    n = len(houses)
    if n == 0:
        return 0
    if n == 1:
        return houses[0]

    dp = [0] * (n + 1)
    dp[1] = houses[0]

    for i in range(2, n + 1):
        dp[i] = max(dp[i - 1], dp[i - 2] + houses[i - 1])
    return dp[n]
getMoney(r"C:\Users\ADMIN\OneDrive\Desktop\10.txt")
```

```python
#12
def maximumProduct(path: str) -> int:
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    n = int(lines[0].strip())
    nums = list(map(int, lines[1].strip().split()))
    nums = nums[:n]
    nums.sort()
    product1 = nums[-1] * nums[-2] * nums[-3]
    product2 = nums[0] * nums[1] * nums[-1]
    return max(product1, product2)
maximumProduct(r"C:\Users\ADMIN\OneDrive\Desktop\12.txt")
```

```python
#13
def move_zeros(path: str):
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    n = int(lines[0].strip())
    arr = list(map(int, lines[1].strip().split()))
    arr = arr[:n]
    non_zero = []
    zeros = []
    for num in arr:
        if num == 0:
            zeros.append(num)
        else:
            non_zero.append(num)
    result = non_zero + zeros
    return result
move_zeros(r"C:\Users\ADMIN\OneDrive\Desktop\13.txt")
```

```python
#14
import csv

def averageTime(path: str) -> float:
    total_time = 0
    lap_count = 0
    with open(path, 'r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        header = next(reader)
        for row in reader:
            if len(row) >= 2:
                try:
                    time_val = float(row[1])
                    total_time += time_val
                    lap_count += 1
                except ValueError:
                    continue
    
    if lap_count == 0:
        return 0.0
    avg = total_time / lap_count
    return round(avg, 2)
averageTime(r"C:\Users\ADMIN\OneDrive\Desktop\14.txt")
```

```python
#15
import csv

def findMovies(path: str) -> list[str]:
    result = []
    with open(path, 'r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        header = next(reader)  
        for row in reader:
            if len(row) >= 3:
                title = row[0].strip()
                try:
                    rating = float(row[1].strip())
                except ValueError:
                    continue  
                description = row[2].strip().lower()
                if rating > 8.0 and "boring" not in description:
                    result.append(title)
    
    return result
findMovies(r"C:\Users\ADMIN\OneDrive\Desktop\15.txt")
```

```python
#16 
import math
def calculate_perimeters(path: str):
    perimeters = []
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if line.startswith("TYPE:"):
            shape_type = line.split(":")[1].strip()
            i += 1
            if i >= len(lines):
                break
            params_line = lines[i].strip()
            
            if shape_type == "SQUARE":
                side = float(params_line.split(":")[1].strip())
                perimeter = 4 * side
                perimeters.append(round(perimeter, 2))
                
            elif shape_type == "RECTANGLE":
                parts = params_line.split()
                width = float(parts[1].strip())
                height = float(parts[3].strip())
                perimeter = 2 * (width + height)
                perimeters.append(round(perimeter, 2))
                
            elif shape_type == "CIRCLE":
                radius = float(params_line.split(":")[1].strip())
                perimeter = 2 * math.pi * radius
                perimeters.append(round(perimeter, 2))
        i += 1
    for p in perimeters:
        print(p)

calculate_perimeters(r"C:\Users\ADMIN\OneDrive\Desktop\16.txt")
```

```python
#17
def productExceptSelf(path: str) -> list[int]:
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    

    n = int(lines[0].strip())

    nums = list(map(int, lines[1].strip().split()))
    

    nums = nums[:n]
    
    res = [1] * n
    
  
    left_product = 1
    for i in range(n):
        res[i] = left_product
        left_product *= nums[i]
    

    right_product = 1
    for i in range(n - 1, -1, -1):
        res[i] *= right_product
        right_product *= nums[i]
    
    return res
productExceptSelf(r"C:\Users\ADMIN\OneDrive\Desktop\17.txt")
```

```python
#18
import math

def estimatedTime(path: str) -> float:
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    
    x1 = y1 = x2 = y2 = v = 0.0
    
    for line in lines:
        line = line.strip()
        if line.startswith("SRC"):
            parts = line.split()
            x1 = float(parts[2])
            y1 = float(parts[4])
        elif line.startswith("DES"):
            parts = line.split()
            x2 = float(parts[2])
            y2 = float(parts[4])
        elif line.startswith("VELOCITY"):
            parts = line.split()
            v = float(parts[1])

    distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

    time = distance / v
    

    return round(time, 2)
estimatedTime(r"C:\Users\ADMIN\OneDrive\Desktop\18.txt")
```

```python
#20
def imageSmoother(path: str):
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    

    M = int(lines[0].strip())
    img = []
    for i in range(1, M + 1):
        row = list(map(int, lines[i].strip().split()))
        img.append(row)

    smoothed = [[0] * M for _ in range(M)]

    directions = [(-1, -1), (-1, 0), (-1, 1),
                  (0, -1),  (0, 0),  (0, 1),
                  (1, -1),  (1, 0),  (1, 1)]
    
    for i in range(M):
        for j in range(M):
            total = 0
            count = 0
            for di, dj in directions:
                ni, nj = i + di, j + dj
                if 0 <= ni < M and 0 <= nj < M:
                    total += img[ni][nj]
                    count += 1
            smoothed[i][j] = total // count  
    
  
    for row in smoothed:
        print(' '.join(map(str, row)))
imageSmoother(r"C:\Users\ADMIN\OneDrive\Desktop\20.txt")
```
