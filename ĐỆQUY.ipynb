```python
def merge_sorted_recursive(list1, list2):
    if not list1:
        return list2
    if not list2:
        return list1
    if list1[0] < list2[0]:
        return [list1[0]] + merge_sorted_recursive(list1[1:], list2)
    else:
        return [list2[0]] + merge_sorted_recursive(list1, list2[1:])

def merge_multiple_sorted_recursive(lists):
    if not lists:
        return []
    if len(lists) == 1:
        return lists[0]
    
    mid = len(lists) // 2
    left = merge_multiple_sorted_recursive(lists[:mid])
    right = merge_multiple_sorted_recursive(lists[mid:])
    return merge_sorted_recursive(left, right)
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
print(merge_sorted_recursive(list1, list2))
```

```python
import heapq

def find_kth_largest(matrix, k):
    if not matrix or k <= 0:
        return None
    
    total = sum(len(row) for row in matrix)
    
    if k > total:
        return None
    
    heap = []
    
    for row in matrix:
        for num in row:
            heapq.heappush(heap, num)
            if len(heap) > k:
                heapq.heappop(heap)
    
    return heap[0]

def main():
    print("NHẬP DANH SÁCH CÁC DÒNG SỐ (bấm Enter để kết thúc nhập):")
    matrix = []
    
    while True:
        line = input().strip()
        if line == "":
            break
        
        try:
            row = list(map(int, line.split()))
            matrix.append(row)
        except:
            print("Dòng không hợp lệ, vui lòng nhập lại")
    
    if not matrix:
        print("Không có dữ liệu đầu vào")
        return
    
    print("\nNhập giá trị k: ")
    try:
        k = int(input().strip())
    except:
        print("k phải là số nguyên")
        return
    
    result = find_kth_largest(matrix, k)
    
    print(f"\nKết quả:")
    print(f"Ma trận: {matrix}")
    print(f"k = {k}")
    
    if result is None:
        if k <= 0:
            print("k phải là số dương")
        else:
            total = sum(len(row) for row in matrix)
            print(f"k = {k} lớn hơn tổng số phần tử ({total})")
    else:
        print(f"Số lớn thứ {k}: {result}")

if __name__ == "__main__":
    main()
```

```python
def solve(n_str):
    s = list(n_str)
    length = len(s)
    
    for i in range(length - 1):
        min_idx = i
       
        for j in range(length - 1, i, -1):
           
            if i == 0 and s[j] == '0':
                continue
            if s[j] < s[min_idx]:
                if min_idx == i or s[j] < s[min_idx]:
                    min_idx = j
        

        if min_idx != i:
            s[i], s[min_idx] = s[min_idx], s[i]
            return "".join(s)
            
    return n_str 
n_str = input()
print(solve(n_str))
```

```python
def is_subsequence(list1, list2):
    n1, n2 = len(list1), len(list2)
    
    if n1 > n2:
        return False
    
    for i in range(n2 - n1 + 1):
        found = True
        for j in range(n1):
            if list1[j] != list2[i + j]:
                found = False
                break
        if found:
            return True
    
    return False

def main():
   

    try:
        list1 = list(map(int, input().split()))
    except:
        print("Đầu vào không hợp lệ!")
        return
    

    try:
        list2 = list(map(int, input().split()))
    except:
        print("Đầu vào không hợp lệ!")
        return
    
    result = is_subsequence(list1, list2)
    print(f" {'YES' if result else 'NO'}")

if __name__ == "__main__":
    main()
```

```python
def can_place_flowers(flowerbed, k):
    n = len(flowerbed)
    count = 0
    
    for i in range(n):
        if flowerbed[i] == 0:
            left_empty = (i == 0) or (flowerbed[i-1] == 0)
            right_empty = (i == n-1) or (flowerbed[i+1] == 0)
            
            if left_empty and right_empty:
                flowerbed[i] = 1
                count += 1
                
                if count >= k:
                    return True
    
    return count >= k

def main():
    flowerbed = list(map(int, input().split()))
    k = int(input())
    
    result = can_place_flowers(flowerbed.copy(), k)
    
    print( f"{'True' if result else 'False'}")
    
   
   
if __name__ == "__main__":
    main()
```

```python
def find_median(nums):
    a, b, c, d, e = nums
    
    if a > b: a, b = b, a
    if c > d: c, d = d, c
    if a > c: a, c = c, a; b, d = d, b
    
    if e > c:
        if e > d:
            return c if b > d else (b if b > d else d)
        else:
            return c if b > e else (b if b > e else e)
    else:
        if e > a:
            return c if b > e else (b if b > e else e)
        else:
            return c if b > c else b


nums = list(map(int, input().split()))
result = find_median(nums)
print(result)
```

```python
def find_radius(houses, heaters):
    houses.sort()
    heaters.sort()
    
    i = 0
    max_dist = 0
    
    for house in houses:
        while i < len(heaters)-1 and abs(heaters[i]-house) >= abs(heaters[i+1]-house):
            i += 1
        max_dist = max(max_dist, abs(heaters[i]-house))
    
    return max_dist

houses = list(map(int, input("Nhà: ").split()))
heaters = list(map(int, input("Lò sưởi: ").split()))
print(find_radius(houses, heaters))
```

```python
def next_permutation(arr):
    n = len(arr)
    i = n - 2
    while i >= 0 and arr[i] >= arr[i + 1]:
        i -= 1
    if i == -1:
        return sorted(arr) 
    j = n - 1
    while arr[j] <= arr[i]:
        j -= 1
    arr[i], arr[j] = arr[j], arr[i]
    left, right = i + 1, n - 1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    return arr
n = int(input())
arr = list(map(int, input().split()))
result = next_permutation(arr)
print(' '.join(map(str, result)))
```
