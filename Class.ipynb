```python
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    def is_valid_date(self):
        if self.month < 1 or self.month > 12:
            return False
        days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        def is_leap_year(year):
            return (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0)
        if self.month == 2 and is_leap_year(self.year):
            max_days = 29
        else:
            max_days = days_in_month[self.month - 1]
        if self.day < 1 or self.day > max_days:
            return False
        return True
    def next_day(self):
        if not self.is_valid_date():
            return "INVALID"
        next_day = self.day
        next_month = self.month
        next_year = self.year
        days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        def is_leap_year(year):
            return (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0)
        if next_month == 2 and is_leap_year(next_year):
            days_in_month[1] = 29
        next_day += 1
        if next_day > days_in_month[next_month - 1]:
            next_day = 1
            next_month += 1
            if next_month > 12:
                next_month = 1
                next_year += 1
        result = f"{next_day:02d}/{next_month:02d}/{next_year:04d}"
        return result
n = input().split('/')
d = int(n[0])
m = int(n[1])
y = int(n[2])
date = Date(d,m,y)
print(date.next_day())
```

```python
import math 
class checkpoint:
  def __init__(self, x, y):
      self.x = x
      self.y = y 
  def kiem_tra_vi_tr(self):
      if self.x == 0 and  self.y == 0:
         return 'M nam trung voi goc toa do'
      elif self.x !=0 and self.y == 0:
         return ' M nam tren truc hoanh'
      elif self.x ==0 and self.y !=0:
          return 'M nam tren truc tung'
      else:
          return 'M nam tren mat phang toa do'
  def tinh_khoang_cach_den_goc(self):
    d = math.sqrt( self.x**2 + self.y**2)
    return d 
a,b = map(float, input().split())
c = checkpoint(a,b)
print(c.kiem_tra_vi_tr())
print(f'{c.tinh_khoang_cach_den_goc():.2f}')
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = []
    
    def them_san_pham(self, ten, gia):
        sp = {"name": ten, "price": gia}
        self.items.append(sp)
        print(f"Da them '{ten}' voi gia {gia}")
    
    def xoa_san_pham(self, ten):
        for sp in self.items:
            if sp["name"] == ten:
                self.items.remove(sp)
                print(f"Da xoa '{ten}'")
                return
        print(f"Khong tim thay '{ten}'")
    
    def kiem_tra_rong(self):
        return len(self.items) == 0
    
    def tong_tien(self):
        tong = sum(sp["price"] for sp in self.items)
        return tong
    
    def hien_thi_gio_hang(self):
        if self.kiem_tra_rong():
            print("Gio hang rong")
        else:
            print("Gio hang:")
            for sp in self.items:
                print(f"  - {sp['name']}: {sp['price']}")
    
    def xoa_toan_bo(self):
        self.items.clear()
        print("Da xoa toan bo gio hang")

# Chương trình chính
cart = ShoppingCart()

while True:
    print("\n=== MENU ===")
    print("1. Hien thi gio hang")
    print("2. Them san pham")
    print("3. Xoa san pham")
    print("4. Kiem tra gio rong")
    print("5. Tinh tong tien")
    print("6. Xoa toan bo gio hang")
    print("0. Thoat")
    
    chon = input("Chon chuc nang (0-6): ")
    
    if chon == '1':
        cart.hien_thi_gio_hang()
    elif chon == '2':
        ten = input("Nhap ten san pham: ")
        gia = float(input("Nhap gia: "))
        cart.them_san_pham(ten, gia)
    elif chon == '3':
        ten = input("Nhap ten san pham can xoa: ")
        cart.xoa_san_pham(ten)
    elif chon == '4':
        if cart.kiem_tra_rong():
            print("Gio hang rong")
        else:
            print("Gio hang co san pham")
    elif chon == '5':
        print(f"Tong tien: {cart.tong_tien()}")
    elif chon == '6':
        cart.xoa_toan_bo()
    elif chon == '0':
        thoat = input("Co muon thoat khong? (yes/no): ")
        if thoat.lower() == "yes":
            print("Cam on!")
            break
    else:
        print("Lua chon khong hop le")
```

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return abs(a)

class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mau so khong the bang 0")
        
        if den < 0:
            num = -num
            den = -den
        
        ucln = gcd(num, den)
        self.num = num // ucln
        self.den = den // ucln
    
    def __str__(self):
        return f"{self.num}/{self.den}"
    
    def add(self, other):
        num = self.num * other.den + other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)
    
    def sub(self, other):
        num = self.num * other.den - other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)
    
    def mul(self, other):
        num = self.num * other.num
        den = self.den * other.den
        return Fraction(num, den)
    
    def div(self, other):
        num = self.num * other.den
        den = self.den * other.num
        return Fraction(num, den)

a, b, op, c, d = input().split()
a, b, c, d = map(int, [a, b, c, d])

frac1 = Fraction(a, b)
frac2 = Fraction(c, d)

if op == '+':
    result = frac1.add(frac2)
elif op == '-':
    result = frac1.sub(frac2)
elif op == '*':
    result = frac1.mul(frac2)
elif op == '/':
    result = frac1.div(frac2)

print(result)
```

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
    
    def deposit(self, x):
        self.balance += x
    
    def withdraw(self, x):
        if x <= self.balance: 
            self.balance -= x
    
    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"

owner, balance = input().split()
balance = float(balance)
account = BankAccount(owner, balance)
n = int(input())
for _ in range(n):
    cmd = input().split()
    if cmd[0] == 'DEPOSIT':
        cmd[1] = float(cmd[1])
        account.deposit(cmd[1])
    elif cmd[0] == 'WITHDRAW':
         cmd[1] = float(cmd[1])
         account.withdraw(cmd[1])
print(account)
```

```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        count = 0
        for book in self.books:
            if book.author == name:
                count += 1
        return count

    def find_by_year(self, y):
        count = 0
        for book in self.books:
          
            if str(book.year) == str(y):
                count += 1
        return count


my_library = Library()


try:
    n = int(input())
except:
    n = 0

for _ in range(n):
    line = input().split(' ', 1)
    cmd = line[0]
    
    if cmd == "ADD":
     
        data = line[1].split(';')
        title = data[0]
        author = data[1]
        year = data[2]
        
        new_book = Book(title, author, year)
        my_library.add(new_book)
        
    elif cmd == "COUNT":
     
        author_name = line[1]
        print(my_library.count_by_author(author_name))
        
    elif cmd == "COUNTYEAR":
  
        year_to_find = line[1]
        print(my_library.find_by_year(year_to_find))
        
```

```python
import math

class Point:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    def __sub__(self, no):
        return Point(self.x - no.x, self.y - no.y, self.z - no.z)
    def dot(self, no):
        return self.x * no.x + self.y * no.y + self.z * no.z

    def cross(self, no):
        return Point(
            self.y * no.z - self.z * no.y,
            self.z * no.x - self.x * no.z,
            self.x * no.y - self.y * no.x
        )
    def absolute(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)

points = []
for i in range(4):
    l = list(map(float, input().split()))
    points.append(Point(l[0], l[1], l[2]))

a, b, c, d = points


ab = b - a
bc = c - b
cd = d - c


n1 = ab.cross(bc)

n2 = bc.cross(cd)


cos_phi = n1.dot(n2) / (n1.absolute() * n2.absolute())


cos_phi = max(-1, min(1, cos_phi))


phi_radian = math.acos(cos_phi)
phi_degree = math.degrees(phi_radian)

print(f"{phi_degree:.2f}")

        
        
```

```python
class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag
    
    def multiply(self, other):
        real_part = self.real * other.real - self.imag * other.imag
        imag_part = self.real * other.imag + self.imag * other.real
        return ComplexNumber(real_part, imag_part)
    
    def display(self):
        print(f"{self.real} {self.imag}")



    r1, i1 = map(float, input().split())
    r2, i2 = map(float, input().split())

    num1 = ComplexNumber(r1, i1)
    num2 = ComplexNumber(r2, i2)
    

    result = num1.multiply(num2)
    

    result.display()
```

```python
import math

class Triangle:

    def __init__(self, initEdge1, initEdge2, initEdge3):
        self.edge1 = initEdge1
        self.edge2 = initEdge2
        self.edge3 = initEdge3

    def getArea(self):
        a, b, c = self.edge1, self.edge2, self.edge3
        
        if (a + b > c) and (a + c > b) and (b + c > a):
            s = (a + b + c) / 2 # Nửa chu vi
            area = math.sqrt(s * (s - a) * (s - b) * (s - c))
            return area
        else:
            return -1

try:
    data = list(map(int, input().split()))
    if len(data) == 3:
        e1, e2, e3 = data
        
     
        my_triangle = Triangle(e1, e2, e3)
        
     
        result = my_triangle.getArea()
        
  
        if result == -1:
            print("invalid")
        else:
          
            print(f"{result:.2f}")
    else:
        print("invalid")
except:
    print("invalid")
```
