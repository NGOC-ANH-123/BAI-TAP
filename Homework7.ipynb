```python
#1
def binary_search(arr, left, right, target):
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

a = list(map(int, input().split()))
t = int(input())
kq = binary_search(a, 0, len(a)-1, t)
print(kq)

    
```

```python
#2
def count_occurrences(arr, x):
    count = 0
    for num in arr:
        if num == x:
            count += 1
    return count

a = list(map(int, input().split()))
x = int(input())
kq = count_occurrences(a, x)
print(kq)
```

```python
#3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

a = list(map(int, input().split()))
kq = bubble_sort(a)
print(' '.join(map(str, kq)))
```

```python
#4
def find_most_frequent(arr):
    freq = {}
    max_count = 0
    result = arr[0]
    
    for num in arr:
        freq[num] = freq.get(num, 0) + 1
        if freq[num] > max_count:
            max_count = freq[num]
            result = num
    return result, max_count

a = list(map(int, input().split()))
kq, cnt = find_most_frequent(a)
print(f"{kq} xuat hien nhieu nhat, som nhat, {cnt} lan")
```

```python
#5
def count_pairs(arr, X):
    count = 0
    seen = {}
    
    for num in arr:
        complement = X - num
        if complement in seen:
            count += seen[complement]
        seen[num] = seen.get(num, 0) + 1
    
    return count

a = list(map(int, input().split()))
x = int(input())
kq = count_pairs(a, x)
print(kq)
```

```python
#6
import ast
def flatten_list(nested_list):
    result = []
    for sublist in nested_list:
        result.extend(sublist)
    return result
s = input().strip()
ds = ast.literal_eval(s)
kq = flatten_list(ds)
print(' '.join(map(str, kq)))
```

```python
#7
def length_of_LIS(nums):
    if not nums:
        return 0
    dp = [1] * len(nums)
    
    for i in range(len(nums)):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    
    return max(dp)

a = list(map(int, input().split()))
kq = length_of_LIS(a)
print(kq)
```

```python
#8
import ast
def min_interval_length(intervals, queries):
    intervals.sort(key=lambda x: x[1] - x[0])
    res = []
    for q in queries:
        ans = -1
        for l, r in intervals:
            if l <= q <= r:
                length = r - l + 1
                if ans == -1 or length < ans:
                    ans = length
        res.append(ans)
    return res

intervals = ast.literal_eval(input().strip())
queries = ast.literal_eval(input().strip())

result = min_interval_length(intervals, queries)
print(' '.join(map(str, result)))
```

```python
#9
def count_smaller(nums):
    result = []
    for i in range(len(nums)):
        count = 0
        for j in range(i + 1, len(nums)):
            if nums[j] < nums[i]:
                count += 1
        result.append(count)
    return result

a = list(map(int, input().split()))
kq = count_smaller(a)
print(' '.join(map(str, kq)))
```

```python
#10
def max_subarray_sum_modulo(nums, m):
    prefix = 0
    max_sum = 0
    prefixes = [0]
    
    for num in nums:
        prefix = (prefix + num) % m
        max_sum = max(max_sum, prefix)
        
        for p in prefixes:
            if p > prefix:
                max_sum = max(max_sum, (prefix - p + m) % m)
        
        prefixes.append(prefix)
        prefixes.sort()
    
    return max_sum

a = list(map(int, input().split()))
m = int(input())
kq = max_subarray_sum_modulo(a, m)
print(kq)
```

```python
#11
def count_word_occurrences(text, keyword):
    words = text.split()
    count = 0
    for word in words:
        if word.strip(".,!?;:\"") == keyword:
            count += 1
    return count

text = input().strip()
keyword = input().strip()
kq = count_word_occurrences(text, keyword)
print(kq)
```

```python
#13
def find_harmonic_subsequence(arr):
    freq = {}
    for num in arr:
        freq[num] = freq.get(num, 0) + 1
    
    max_length = 0
    for num in freq:
        if num + 1 in freq:
            max_length = max(max_length, freq[num] + freq[num + 1])
    
    return max_length if max_length > 0 else 0

arr = list(map(int, input().split()))
result = find_harmonic_subsequence(arr)
print(result)
```

```python
#14
def find_sevens_in_array(arr):
    positions = []
    for i in range(len(arr)):
        if arr[i] == 7:
            positions.append(i)
    
    if not positions:
        return "Not found"
    
    positions.sort(reverse=True)
    return " ".join(map(str, positions))

n = int(input())
arr = list(map(int, input().split()))

result = find_sevens_in_array(arr)
print(result)
```

```python
#15
def find_neighbors(names):
    try:
        index = names.index("Nemo")
        if index == 0:
            return f"{names[-1]} and {names[1]}"
        elif index == len(names) - 1:
            return f"{names[index-1]} and {names[0]}"
        else:
            return f"{names[index-1]} and {names[index+1]}"
    except ValueError:
        return "Nemo not found"

n = int(input())
names = []
for _ in range(n):
    names.append(input().strip())

kq = find_neighbors(names)
print(kq)
```
