```python
def find_max_position(lst):
    if not lst:
        return -1
    max_val = lst[0]
    max_pos = 0
    for i in range(1, len(lst)):
        if lst[i] > max_val:
            max_val = lst[i]
            max_pos = i
    return max_pos
lst = list(map(int, input().split()))
print(find_max_position(lst))
```

```python
def find_first_k(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1
lst = list(map(int, input().split()))
k = int(input())
print(find_first_k(lst, k))
```

```python
words = input().split()
d = {}
for idx, word in enumerate(words):
    if word not in d:
        d[word] = []
    d[word].append(idx)
print(d)
```

```python
def remove_less_than_x(lst, x):
    i = 0
    while i < len(lst):
        if lst[i] < x:
            lst.pop(i)
        else:
            i += 1
    return lst
lst = list(map(int, input().split()))
x = int(input())
print(remove_less_than_x(lst, x))
```

```python

def move_zeros_to_end(lst):
    non_zeros = [v for v in lst if v != 0]
    zeros = [v for v in lst if v == 0]
    return non_zeros + zeros
lst = list(map(int, input().split()))
print(move_zeros_to_end(lst))
```

```python
def missing_values(lst):
    all_values = set(range(10))
    present = set(lst)
    missing = sorted(all_values - present)
    return missing
lst = list(map(int, input().split()))
print(missing_values(lst))
```

```python
def sum_local_max(lst):
    total = 0
    for i in range(1, len(lst)-1):
        if lst[i] > lst[i-1] and lst[i] > lst[i+1]:
            total += lst[i]
    return total
lst = list(map(int, input().split()))
print(sum_local_max(lst))
```

```python
def classify_sequence(lst):
    increasing = all(lst[i] < lst[i+1] for i in range(len(lst)-1))
    decreasing = all(lst[i] > lst[i+1] for i in range(len(lst)-1))
    if increasing:
        return "Tăng thực sự"
    elif decreasing:
        return "Giảm thực sự"
    else:
        return "Vô danh"
lst = list(map(int, input().split()))
print(classify_sequence(lst))
```
