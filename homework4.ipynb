```python
# Bai 4 
h = int(input("Nhập chiều cao tam giác: "))

for i in range(1, h):
    print(" " * (h - i), end="")
    if i == 1:
        print("*")
    else:
        print("*" + " " * (2 * i - 3) + "*")

print("*" * (2 * h - 1))  # Dòng cuối cùng
```

```python
# Bai 14 
so_lon_nhat = float('-inf')
so_nho_nhat = float('inf')

while True:
    so = int(input("Nhập một số nguyên (-1 để kết thúc): "))
    if so == -1:
        break
    if so > so_lon_nhat:
        so_lon_nhat = so
    if so < so_nho_nhat:
        so_nho_nhat = so

print("Số lớn nhất là:", so_lon_nhat)
print("Số nhỏ nhất là:", so_nho_nhat)
```

```python
# Bai 16
n = int(input("Nhập số nguyên dương n: "))
so_chan = 0
so_le = 0

while n > 0:
    chu_so = n % 10  # Lấy chữ số cuối cùng
    if chu_so % 2 == 0:
        so_chan += 1
    else:
        so_le += 1
    n = n // 10  # Bỏ chữ số cuối cùng
print("Số chữ số chẵn là:", so_chan)
print("Số chữ số lẻ là:", so_le)
```

```python
# Bai 18 cach 1 
n = int(input("Nhập số nguyên dương: "))

if n < 1:
    print("Không phải lũy thừa của 3")
else:
    while n % 3 == 0:
        n = n // 3
    if n == 1:
        print("Đúng! Là lũy thừa của 3")
    else:
        print("Không phải lũy thừa của 3")
```

```python
# Bai 19 
A = int(input("Nhập số nguyên dương A: "))

f1 = 1
f2 = 1
while f2 <= A:
    f1, f2 = f2, f1 + f2
    print(f2)

print("Số Fibonacci lớn nhất không vượt quá A là:", f1)
```

```python
# Bai 20 
n = int(input("Nhập số nguyên dương n: "))
dem = 0

while n > 0:
    chu_so = n % 10
    if chu_so in (2, 3, 5, 7):
        dem += 1
    n = n // 10

print("Số chữ số nguyên tố là:", dem)

```
