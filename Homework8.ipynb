```python
#1
from collections import Counter
def solve():
    arr = list(map(int, input().split()))
    freq = Counter(arr)
    
    delete_count = 0
    for num in list(freq.keys()):
        count = freq[num]
        if num == 0:
            if count > 0:
                delete_count += count
            del freq[num]
        else:
            if count < num:
                delete_count += count
                del freq[num]
            elif count > num:
                delete_count += count - num
                freq[num] = num
    
    if delete_count == 0:
        print("YES")
    else:
        print(f"NO {delete_count}")

solve()
```

```python
#2
def solve():
    arr = list(map(int, input().split()))
    if not arr:
        return
    
    cand = None
    count = 0
    
    for num in arr:
        if count == 0:
            cand = num
        if num == cand:
            count += 1
        else:
            count -= 1
    
    if arr.count(cand) > len(arr) // 2:
        print(cand)

solve()
```

```python
#3
arr = list(map(int, input().split()))
even = []
odd = []

for num in arr:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)

even.sort()
odd.sort(reverse=True)

result = even + odd
print(' '.join(map(str, result)))
```

```python
#4
a = list(map(int, input().split()))
b = list(map(int, input().split()))

def is_ascending(arr):
    return arr[0] < arr[-1]

if is_ascending(a) == is_ascending(b):
    if a[0] < a[-1]:
        a.extend(b)
        a.sort()
    else:
        a.extend(b)
        a.sort(reverse=True)
    print(' '.join(map(str, a)))
else:
    if a[0] < a[-1]:
        for num in b:
            a.append(num)
        a.sort()
    else:
        for num in b:
            a.append(num)
        a.sort(reverse=True)
    print(' '.join(map(str, a)))
    
```

```python
#7
n = int(input())
first_ten = ""
count_ten = {}

for _ in range(n):
    full_name = input().strip()
    ten = full_name.split()[-1]
    if not first_ten:
        first_ten = ten
    if ten in count_ten:
        count_ten[ten] += 1
    else:
        count_ten[ten] = 1

max_count = 0
best_ten = first_ten
for ten, cnt in count_ten.items():
    if cnt > max_count:
        max_count = cnt
        best_ten = ten

print(best_ten)
```

```python
#8
s = input().strip()
dem = 0

for word in s.split():
    word = word.strip('.,!?;:')
    
    if len(word) < 2:
        continue
    
    if not word[0].isupper():
        continue
    
    ok = True
    for c in word[1:]:
        if not c.islower():
            ok = False
            break
    
    if ok and word.isalpha():
        dem += 1

print(dem)
```

```python
#12
s = input().strip()
stack = []
ok = True
pairs = {')': '(', ']': '[', '}': '{', '>': '<'}

for c in s:
    if c in '([{<':
        stack.append(c)
    elif c in ')]}>':
        if not stack or stack[-1] != pairs[c]:
            ok = False
            break
        stack.pop()

if stack:
    ok = False

print("YES" if ok else "NO")
```

```python
#12
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

n = int(input())

max_len = 0
best_start = 0
current_len = 0
current_start = 0

for i in range(1, n + 1):
    if not is_prime(i):
        if current_len == 0:
            current_start = i
        current_len += 1
        if current_len > max_len:
            max_len = current_len
            best_start = current_start
    else:
        current_len = 0

result = []
for j in range(max_len):
    result.append(best_start + j)

print(', '.join(map(str, result)))
```

```python
#14
arr = list(map(int, input().split()))
n = len(arr)
if n == 0:
    print(0)
    exit()

dp = [1] * n
for i in range(n):
    for j in range(i):
        if arr[i] > arr[j]:
            dp[i] = max(dp[i], dp[j] + 1)

max_len = max(dp)
result = []
curr_len = max_len
for i in range(n-1, -1, -1):
    if dp[i] == curr_len:
        result.append(arr[i])
        curr_len -= 1
        if curr_len == 0:
            break

result.reverse()
print(' '.join(map(str, result)))
```
