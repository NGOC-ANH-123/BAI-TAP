```python
# Kiểm tra số nguyên tố
while True:
    n = int(input("Nhập số nguyên dương n: "))
    if n > 0:
        break

if n < 2:
    print("Không phải số nguyên tố")
else:
    dem = 0
    for i in range(2, n):
        if n % i == 0:
            dem += 1
    if dem == 0:
        print(n, "là số nguyên tố")
    else:
        print(n, "không phải là số nguyên tố")
```

```python
# Đếm số chữ số của n (không tính dấu)
n = int(input("Nhập số nguyên n: "))
if n < 0:
    n = -n

dem = 0
if n == 0:
    dem = 1
else:
    while n > 0:
        n //= 10
        dem += 1

print("Số chữ số là:", dem)
```

```python
# Tính tổng các số nguyên tố trong đoạn [a, b]
a = int(input("Nhập a: "))
b = int(input("Nhập b: "))
tong = 0
for i in range(a, b + 1):
    if i >= 2:
        dem = 0
        for j in range(2, i):
            if i % j == 0:
                dem += 1
        if dem == 0:
            tong += i
print("Tổng các số nguyên tố từ", a, "đến", b, "là:", tong)
```

```python
# Tìm ước số nguyên tố lớn nhất của n
n = int(input("Nhập n (n ≥ 2): "))
uoc = 2
while n > 1:
    if n % uoc == 0:
        n //= uoc
    else:
        uoc += 1
print("Ước số nguyên tố lớn nhất là:", uoc)
```

```python
# Lặp n = n + reverse(n) cho đến khi n là palindrome
n = int(input("Nhập n: "))
buoc = 0
while True:
    m = n
    dao = 0
    while m > 0:
        dao = dao * 10 + m % 10
        m //= 10
    if n == dao:
        break
    n = n + dao
    buoc += 1
print("Số bước:", buoc)
print("Kết quả palindrome:", n)
```

```python
# In các số chính phương <= n có chữ số đôi một khác nhau
n = int(input("Nhập n: "))
found = False
for i in range(1, n + 1):
    sq = i * i
    if sq > n:
        break
    x = sq
    hop = True
    for d in range(10):
        dem = 0
        m = x
        while m > 0:
            if m % 10 == d:
                dem += 1
            m //= 10
        if dem > 1:
            hop = False
            break
    if hop:
        print(sq, end=" ")
        found = True
if not found:
    print("no number")
```

```python
# Đếm số lượng ước số chẵn của n
n = int(input("Nhập n: "))
dem = 0
for i in range(1, n + 1):
    if n % i == 0 and i % 2 == 0:
        dem += 1
print("Số lượng ước số chẵn là:", dem)
```

```python
# Kiểm tra cặp số thân thiết
a = int(input("Nhập số a: "))
b = int(input("Nhập số b: "))

tong1 = 0
for i in range(1, a):
    if a % i == 0:
        tong1 += i

tong2 = 0
for i in range(1, b):
    if b % i == 0:
        tong2 += i

if tong1 == b and tong2 == a:
    print("true")
else:
    print("false")
```

```python
# Tìm ước chung lớn nhất (UCLN)
m = int(input("Nhập m: "))
n = int(input("Nhập n: "))
while m != n:
    if m > n:
        m -= n
    else:
        n -= m
print("Ước chung lớn nhất là:", m)
```

```python
# Bài toán gà và chó
con = int(input("Nhập tổng số con: "))
chan = int(input("Nhập tổng số chân: "))

tim = False
for ga in range(con + 1):
    cho = con - ga
    if 2 * ga + 4 * cho == chan:
        print("Gà:", ga)
        print("Chó:", cho)
        tim = True
        break
if not tim:
    print("invalid")
```
