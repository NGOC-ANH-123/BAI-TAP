```python
#1
def binary_search(arr, value):
    left , right = 0 , len(arr) - 1
    mid = (left + right )//2
    if arr[mid] == value:
        return mid 
    elif arr[mid] < value:
        left = mid + 1
    else:
        right = mid - 1
    return - 1
arr = list(map(int, input().split()))
value = int(input())
print(binary_search(arr, value))
```

```python
#2
def count_occurence(arr,k):
    s = 0 
    for i in range(len(arr)):
        if arr[i] == k:
            s = s + 1
    return s
arr = list(map(int, input().split()))
k = int(input())
print(count_occurence(arr,k))
```

```python
#3
def bublle_sort(arr):
    n = len(arr)
    for i in range (n):
        swapped = False 
        for j in range(0,n-i-1):
            if arr[j] > arr[j+1]:
                arr[j] , arr[j+1] = arr[j+1] , arr[j]
                swapped = True 
        if not swapped:
            break
    return arr 
arr = list(map(int, input().split()))
print(bublle_sort(arr))
    
        
```

```python
#4
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i 
        for j in range(i+1,n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i] , arr[min_idx] = arr[min_idx], arr[i]
    return arr
arr = list(map(int, input().split()))
print(selection_sort(arr))
```

```python
#5
def max_occurence(arr):
    dic = {}
    for num in arr:
        if num not in dic:
            dic[num] = 1 
        else:
            dic[num] += 1 
    max_count = 0 
    for key , value in dic.items():
        if value > max_count:
            max_count  = value 
    result = []
    for key, value in dic.items():
        if dic[key] == max_count:
            result.append(key)
    if not result:
        return "nothing"
    return f'{result[0]} xuat hien nhieu nhat, som nhat {max_count} lan'
arr = list(map(int, input().split()))
print(max_occurence(arr))
```

```python
#6
def count_couple(arr,c):
    s = 0
    for i in range(len(arr)):
        for j in range( i+1 , len(arr)):
            if arr[i] + arr[j] == c:
                s = s + 1
    return s
arr = list(map(int, input().split()))
c = int(input())
print(count_couple(arr,c))
```

```python
#7
import ast 
def flatten_list(lst):
    result = []
    for sublist in lst:
        for item in sublist:
            result.append(item)
    return result 
string = input()
lst = ast.literal_eval(string)
print(flatten_list(lst))
```

```python
#9
import ast

dict1 = ast.literal_eval(input().strip())
dict2 = ast.literal_eval(input().strip())

merged = {}

for key, val in dict1.items():
    merged[key] = val

for key, val in dict2.items():
    if key in merged:
        merged[key] += val
    else:
        merged[key] = val

sorted_keys = sorted(merged.keys())
result = {}
for key in sorted_keys:
    result[key] = merged[key]

print(result)
```

```python
#10
arr = list(map(int, input().split()))
k = int(input())

result = []
n = len(arr)

for i in range(n):
    for j in range(i+1, n):
        if arr[i] + arr[j] == k:
            result.append((arr[i], arr[j]))

result.sort()
print(result)
```

```python
#12
arr = list(map(int, input().split()))
d = {}
for x in arr:
    if x in d:
        d[x] += 1
    else:
        d[x] = 1

best_num = arr[0]
best_cnt = d[arr[0]]

for num, cnt in d.items():
    if cnt > best_cnt or (cnt == best_cnt and num < best_num):
        best_cnt = cnt
        best_num = num

print(best_num)
```
